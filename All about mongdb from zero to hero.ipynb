## Querying to JSON Documents

```python
# get top record in collection in mongodb and tabel in SQL 

collection.find_one()
```

```python
# retreive all data from my collection in School database
# in SQl    ----   select * from Mytabel_collection

for i in collection.find():
    print(i)
```

```python
# query Documents based on specific key value and get that all record that has this key value

# In SQL    ---- select * from Mytable_collection where Name = Maxmell

for i in collection.find({"Name":"Maxmell"}):
    print(i)
```

```python

for i in collection.find({"Name":"fatih"}):
    print(i)
```

```python

for i in collection.find({"Name":"shan"}):
    print(i)
```

```python
# Using Query Operator (  $in,  $lt, $gt )

for i in collection.find({ "qaulification":{"$in":["phd", "masters"]} }):
    print(i)
```

```python
# And and Query ( $lt mean less than )

for i in collection.find({"qaulification":"masters", "Age":{"$lt":26}}):
    print(i)
```

```python
# retrive any record that atleast fullfill any of this condition 
for i in collection.find({"$or":[ {"Name":"farid"},{"qaulification":"masters"} ]}):
    print(i)
```

```python
# retive record that only fullfill these condition otherwise ot it willnot give this.
for i in collection.find({"$and":[ {"Name":"farid"},{"qaulification":"phd"} ]}):
    print(i)
```

# ----Updating Records in Mongodb----

```python
inventory=db.newtable # creating new collection 

inventory.insert_many([
    {"item": "canvas",                     # these are called documents 
     "qty": 100,
     "size": {"h": 28, "w": 35.5, "uom": "cm"},
     "status": "A"},
    {"item": "journal",
     "qty": 25,
     "size": {"h": 14, "w": 21, "uom": "cm"},
     "status": "A"},
    {"item": "mat",
     "qty": 85,
     "size": {"h": 27.9, "w": 35.5, "uom": "cm"},
     "status": "A"},
    {"item": "mousepad",
     "qty": 25,
     "size": {"h": 19, "w": 22.85, "uom": "cm"},
     "status": "P"},
    {"item": "notebook",
     "qty": 50,
     "size": {"h": 8.5, "w": 11, "uom": "in"},
     "status": "P"},
    {"item": "paper",
     "qty": 100,
     "size": {"h": 8.5, "w": 11, "uom": "in"},
     "status": "D"},
    {"item": "planner",
     "qty": 75,
     "size": {"h": 22.85, "w": 30, "uom": "cm"},
     "status": "D"},
    {"item": "postcard",
     "qty": 45,
     "size": {"h": 10, "w": 15.25, "uom": "cm"},
     "status": "A"},
    {"item": "sketchbook",
     "qty": 80,
     "size": {"h": 14, "w": 21, "uom": "cm"},
     "status": "A"},
    {"item": "sketch pad",
     "qty": 95,
     "size": {"h": 22.85, "w": 30.5, "uom": "cm"},
     "status": "A"}])
```

### Update one records

```python
# before updating
for i in inventory.find({"item":"sketch pad"}):
    print(i)
```

```python
# updating and see after updating
inventory.update_one(
{"item":"sketch pad"},
{"$set":{"size.uom":"m","status":"P"},   # changed these values
"$currentDate":{"lastModified":True}}    # added new key value
)

for i in inventory.find({"item":"sketch pad"}):
    print(i)
```

### Update Many records

```python
# before updating
for i in inventory.find({"item":"sketch pad"}):
    print(i)
```

```python
# updating in all record that contain these key values
inventory.update_many(
    {"qty": {"$lt": 50}},   # those records where quality is less than 50 
    {"$set": {"size.uom": "in", "status": "P"},  # change these values upon fulfilling above condition
     "$currentDate": {"lastModified": True}})   # add this also 
# lets see changes in whole record
for i in inventory.find({"qty": {"$lt": 50}}):  # show only those records thats has qlty les than 50
    print(i)
```

###  Replace  one record

```python
# check before we replace
for i in inventory.find({"item": "paper"}):
    print(i)
```

```python
# now we will replace
inventory.replace_one(
    {"item": "paper"},
    {"item": "paper",
     "instock": [
         {"warehouse": "A", "qty": 60},
         {"warehouse": "B", "qty": 40}]})
```

```python
# checking again to see the changes
for i in inventory.find({"item": "paper"}):
    print(i)
```
